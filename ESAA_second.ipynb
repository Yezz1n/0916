###dataframe**<->list<->dictionary<->ndarray 상호변환 가능

*dataframe은 리스트, ndarray와 달리 칼럼명 가지고 있음->데이터 핸들링 가능




````python
import numpy as np


#1차원 형태
col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:',array1.shape)

#리스트를 이용해 dataframe 생성
df_list1=pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n',df_list1)


#넘파이 ndarray를 이용해 dataframe 생성
df_array1=pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n',df_array1)



#2차원 형태

#3개의 칼럼명이 필요함
col_name2=['col1','col2','col3']

#2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 Dataframe으로 변환

list2=[[1,2,3,],
       [4,5,6]]
array2=np.array(list2)
print('array2 shape:',array2.shape)
df_list2=pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n',df_list2)
df_array2=pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n:',df_array2)
````

* 원본 DataFrame에서 드롭된 DataFrame을 다시 원본 DataFrame 객체 변수로 할당되면 원본 DataFrame에서 드롭된 결과를 적용할 경우와 같음(단, 기존 원본 DataFrame 객체 변수는 메모리에서 추후 제거됨)
 
###Index 객체
:판다스의 index 객체는 RDBMS의 PK와 유사하게 DataFrame, Series의 레코드를 고유하게 식별하는 객체

* dataframe,series에서 index 객체만 DataFrame,index 또는 Series,index 속성을 통해 가능

* titanic_df DataFrame에서 index 객체를 추출

* index 객체는 식별성 데이터를 1차원 array로 가지고 있다. 
 * ndarray와 유사하게 단일 값 반환 및 슬라이싱도 가능
 * 한 번 만들어진 DataFrame 및 Series의 index 객체는 함부로 변경할 수 없다

* series 객체는 index 객체를 포함하지만 series 객체에 연산 함수를 적용할 때 index
->index는 오직 식별용으로만 사용

* dataframe 및 series에 reset_index() 메서드를 수행하면 새롭게 인덱스를 연속 숫자 형으로 할당하며 기존 인덱스는  index라는 새로운 칼럼 명으로 추가



````python
titanic_df=pd.read_csv('titnaic_train.csv')
indexes=titanic_df.index
print(indexes)

print('index 객체 array값:\n',indexes.values)

#새롭게 index 추가
titanic_reset_df=titanic_df.reset_index(inplace=False)
titanic_reset_df.head(3)
````

*reset_index(): 인덱스가 연속된 int 숫자형 데이터가 아닐 경우에 다시 이를 연속 int 숫자형 데이터로 만들 때 주로 사용

->series에 reset_index()를 적용하면 series가 아닌 DataFrame이 반환
 + 기존 인덱스가 칼럼으로 추가돼 컬럼이 2개가 되므로 series가 아닌 DataFrame이 반환

 + reset_index()의 parameter 중 drop=True로 설정 시 기존 인덱스는 새로운 칼럼으로 추가되지 않고 삭제(drop)
->그대로 series로 유지


````python
print('###before reset_index###')
value_counts=titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:',type(value_counts))
new_value_counts=value_counts.reset_index(inplace=False)
print('###after reset_index###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:',type(new_value_counts))
````

####명칭 기반 인덱싱과 위치 기반 인덱싱의 구분

만약 data_df의 인덱스 값이 문자열이 아닌 숫자 값(0,1,2,,,)라면 data_df.ix[0,:]의 행 위치값 0은 인덱스 값이 0인지, 첫번째 행 위치인지 헷갈리게 됨
  


````python
data_df_reset=data_df.reset_index()
data_df_reset=data_df_reset.rename(columns={'index':'old_index'})

data_df_reset.index=data_df_reset.index+1
data_df_reset
````

->인덱스가 integer 형인 경우 ix[0,1]은 행 위치 값으로 위치 기반 인덱스값을 사용하는 것이 아니라 dataframe의 인덱스 값을 사용하는데, 인덱스 값이 0인 key는 찾을 수 없기 때문

->ix[1,1]을 적용해야 첫번째 행(인덱스값 1)의 2번째 열 위치에 있는 데이터값을 반환

명칭 기반과 위치 기반의 데이터 형이 서로 같을 경우 명칭 기반이 우선한다고 이해!

####DataFrame loc[] 연산자
:loc[]은 명칭 기반으로 데이터 추출

* 열 위치에는 칼럼명

*loc[]에 슬라이싱 기호':' 적용시 종료 값 -1이 아니라 종료값까지 포함하는 것을 의미


````python
data_df.loc['one','Name']
data_df_reset.loc[1,'Name']

#print('명칭 기반 ix slicing \n',data_df.ix['one':'two','Name'],'\n')
print('위치 기반 iloc slicing \n',data_df.iloc[0:1,0],'\n') #번째 행으로 하나 출력
print('명칭 기반 loc slicing \n',data_df.loc['one':'two','Name'])
````

####groupby() 적용

DataFrame에 groupby() 호출 시 DataFrameGroupBy라는 또 다른 형태의 DataFrame을 반환

* SQL의 groupby와의 차이점
 * dataframe의 groupby() 호출해 반환된 결과에 aggregation함수를 호출 시 groupby() 대상 칼럼을 제외한 모든 칼럼에 해당 aggregation 함수를 적용
 * SQL은 서로 다른 agrregation 함수 적용 시 select절에 나열하기만 하면 되지만, 
 DataFrame은 여러 개의 aggregation 함수명을 DataFrameGroupBy 객체의 agg() 내에 인자로 입력해서 사용
 * SQL은 여러개의 칼럼에 서로 다른 aggregation 함수 groupby 호출 시 각각 적용시켜 쉽지만 DataFrame의 GroupBy() 더 복잡=>딕셔너리 형태

###apply lambda 식으로 데이터 가공
:apply함수에 lambda 식 결합해 DataFrame이나 Series의 레코드 별로 데이터를 가공하는 기능 제공 

> lambda: return 같은 문법 생략하고 함수 내 처리를 한 줄의 식으로 쉽게 변환하는 식

*lambda에서 여러 개의 값을 입력 인자로 사용해야 할 경우에는 map() 사용

````python
lambda_square=lambda x:x**2
print('3의 제곱은:',lambda_square(3))

a=[1,2,3]
squares=map(lambda x:x**2,a)
list(squares)
````

````python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x:len(x))
titanic_df[['Name','Name_len']].head(3)
````

````python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
#반환값이 식보다 먼저!
````

else if가 많이 나와야 하는 경우나 switch case 문의 경우는 else를 내포해서 쓰기보다  별도의 함수 지정이 낫다

````python
def get_category(age):
  cat=''
  if age<=5: cat='Baby'
  elif age<=12:cat ='Child'
  elif age<=18:cat='Teenager'
  elif age<=25:cat='Student'
  elif age<=35:cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'

  return cat

titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x:get_category(x))
titanic_df[['Age','Age_cat']].head()
````
