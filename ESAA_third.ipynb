#### 다변량 그래프 그리기

:3개 이상의 변수를 사용한 그래프

-필요한 이유: 3개 이상의 변수를 사용하는 다변량 그래프는 적재적소에 맞는 그래프 요소를 추가하여 그래프를 표현해야 한다.


##### 다변량 데이터로 다변량 그래프 그리기- 산점도

1. 앞의 산점도 그래프에 성별을 새 변수로 추가

성별은 산점도 그래프에서 색상으로 구분하면 적당

->but, female, male과 같은 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없다.

<font color='blue'> 0,1과 같은 정수를 사용해야 함-> 각 문자열을 정수로 치환하는 함수를 생성해야함

여성은 0 반환하고 남성은 1 반환하는 함수 생성하기

```python
def recode_sex(sex):
  if sex=='Female':
    return 0
  else: return 1
```

2. recode_sex에서 반환된 값(0,1)을 데이터프레임에 추가

->recode_sex 함수가 반환한 값(0,1)을 데이터프레임에 sex_color열을 새로 생성하여 추가:apply 사용

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

3. 테이블당 인원 수(size)도 산점도 그래프에  추가하기

테이블당 인원 수는 점의 크기로 표현
: scatter 메서드에 s,c 인잣값으로 테이블 당 인원 수와 성별의 치환값을 전달


```python
scatter_plot=plt.figure()
axes1=scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10, #size
    c=tips['sex_color'], #color
    alpha=0.5 #투명도
)
axes1.set_title('Total Bill vs Tip colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

* 산점도 그래프 그리기 위해서는 seaborn 라이브러리의 regplot 메서드를 사용

  * regplot  메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있다.

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Scatter plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips,fit_reg=False)
ax.set_title('Scatter plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

2. 산점도 그래프와 히스토그램을 한 번에 그려주는 jointplot 메서드를 사용
x,y 인자에 원하는 열 이름을 지정하고 data 인잣값으로 데이터프레임을 지정하면 그래프를 그릴 수 있다.
* jointplot 메서드도 그래프 이름과 x,y축 이름을 추가할 수 있다.

```python
joint=sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

3. 산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점 존재

if,산점도 그래프의 데이터를 쉽게 구분하고 싶다면 hexion 육각 그래프 이용
* 육각 그래프: 2차원 표면에 육각형 데이터를 쌓아 표현하는 그래프
 * 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현
 
 > jointplot 메서드 그대로 사용하되 hex로 지정


```python
hexbin=sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
hexbin.set_axis_labels(xlabel="Total Bill",ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

4.이차원 밀집도 그리기

:kdeplot 메서드 이용

* shade=True이면 그래프에 음영 효과를 줄 수 있다.

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
               data2=tips['tip'],
               shade=True)
ax.set_title('Kernel Destiny Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

> 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체

* map_upper:대각선을 기준으로 위쪽에 그림 그래프 지정
* map_lower: 대각선을 기준으로 아래쪽에 그릴 그래프 지정
* map_diag: 대각선 중심으로 그래프 그림


```python
#이차원 밀집도 대각선 기준으로 아래쪽, 산점도 그래프 대각선 기준으로 위쪽, 히스토그램과 밀집도는 대각선을 중심으로 

pair_grid=sns.PairGrid(tips)
pair_gird=pair_grid.map_upper(sns.regplot) #t산점도
pair_grid=pair_grid.map_lower(sns.kdeplot) #이차원 밀집도그래프
pair_grid=pair_grid.map_diag(sns.distplot,rug=True) #히스토그램과 밀집도
plt.show()
```

#seaborn 라이브러리로 그래프 스타일 설정하기
:set_style 활용
> 스타일 적용할 부분에서 한 번만 실행하면 된다.
-> 이후 그래프가 모두 지정한 스타일로 바뀜

```python
#그래프에 스타일 적용하기
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

> whitegird 스타일 설정:그래프의 배경에 가로줄이 생김


```python
sns.set_style('whitegrid')
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

>for 문 적용하여 모든 스타일을 하나씩 적용한 그래프

```python
fig=plt.figure()
seaborn_styles=['darkgrid','whitegrid','dark','white','ticks']
for idx,style in enumerate(seaborn_styles):
  plot_position=idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin=sns.violinplot(x='time',y='total_bill',data=tips,ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
